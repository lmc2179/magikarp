# 1) Quadratic regression

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
lr = LinearRegression()
sample_size = len(cooling_consts)
polynomial_data = np.array([[1, c, c**2, c**3] for c in cooling_consts])
interval = np.arange(LOWER, UPPER, 0.01)
polynomial_interval = np.array([[1, c, c**2, c**3] for c in interval])
lr.fit(polynomial_data, sa_costs)
plt.plot(cooling_consts, sa_costs, linewidth=0.0, marker='o')
predicted_values= lr.predict(polynomial_interval)
plt.plot(interval, predicted_values, linewidth=1.0)
plt.plot([LOWER, UPPER], [baseline_cost, baseline_cost])
plt.show()
```
